# Creating CSV data files from parquet files

Next, we will perform a series of manipulations with the data provided in the novel compressed parquet format.

In summary, these are the steps we will follow:


- First, we will load the files into a dataset
- Second, we will calculate the apparent power based on two columns (real power and reactive power)
- Third, we will remove the columns we do not need
- Then, we will reorder the columns in the order we need
- Finally, we will save the obtained dataset in a CSV file

## Generating the data file from the fan.csv file

```python

import pandas as pd
import numpy as np

with open(r'.\data\ozm1\ozm_fan_200MS.parquet', 'rb') as f:
    df_datos_agregados = pd.read_parquet(f, engine='pyarrow')
    

df_datos_agregados.rename(columns={'Time':'timestamp','Active_Power':'W','Reactive_Power':'VAR','Apparent_Power':'VA'} ,inplace=True)
df_datos_agregados.rename(columns={'Frequency':'f','Voltage':'VLN','Power_Factor':'PF','Current':'A' },inplace=True)



df_datos_agregados.drop (['FirstTimestamp','OriginalTimestamp','Tag'],axis = 'columns', inplace=True)
df_datos_agregados.drop (['Voltage_Harmonics_Complex','Current_Harmonics_Complex','Power_Harmonics'],axis = 'columns', inplace=True)
df_datos_agregados.drop (['Flag','Unbalance','Phi'],axis = 'columns', inplace=True)



df_datos_agregados=df_datos_agregados [['timestamp','W','VAR','VA','f','VLN','PF','A']]

print (df_datos_agregados)

df_datos_agregados.to_csv('fan.csv')


```

## Removing dataset indices

In the data files, the auto-incremental value is not useful as it is not an electrical variable. Remember that this value is used as the dataset index.

For this, we will use the timestamp value as the index, as this value is unique.

We will reconstruct the CSV files by removing this field.

### Main counter file

Let's start by removing the index in the aggregation file (i.e., the global counter file)

```python
import pandas as pd
import numpy as np
with open(r'.\data\ozm1\ozm_aggregation_200MS.parquet', 'rb') as f:
    df_datos_agregados = pd.read_parquet(f, engine='pyarrow')
    
df_datos_agregados.rename(columns={'Time':'timestamp','Active_Power':'W','Reactive_Power':'VAR','Apparent_Power':'VA'} ,inplace=True)
df_datos_agregados.rename(columns={'Frequency':'f','Voltage':'VLN','Power_Factor':'PF','Current':'A' },inplace=True)



df_datos_agregados.drop (['FirstTimestamp','OriginalTimestamp','Tag'],axis = 'columns', inplace=True)
df_datos_agregados.drop (['Voltage_Harmonics_Complex','Current_Harmonics_Complex','Power_Harmonics'],axis = 'columns', inplace=True)
df_datos_agregados.drop (['Flag','Unbalance','Phi'],axis = 'columns', inplace=True)



df_datos_agregados=df_datos_agregados [['timestamp','W','VAR','VA','f','VLN','PF','A']]



df_reset=df_datos_agregados.set_index('timestamp')


print (df_reset)

df_reset.to_csv('aggregation.csv')
```

```python
import pandas as pd

with open(r'.\data\ozm1\ozm_aggregation_200MS.parquet', 'rb') as f:
    df_datos_agregados = pd.read_parquet(f, engine='pyarrow')

print ('filas fichero aggregation.hdf5=',len(df_datos_agregados.index))
```

```python
N=1
df_datos_agregados.head(N)
```

```python
FirstTimestamp=df_datos_agregados.iat[0,1]
```

```python
OriginalTimestamp=df_datos_agregados.iat[0,2]
```

```python
Time=df_datos_agregados.iat[0,3]
```

```python
import pandas as pd
import numpy as np
import datetime
df=pd.DataFrame()

i=[FirstTimestamp,OriginalTimestamp,Time]
df['date']=pd.Series(i)
df['date']=df['date'].apply(lambda d: datetime.datetime.fromtimestamp(int(d)/1000).strftime('%Y-%m-%d %H:%M:%S'))

df

```
